```python
import requests
from enum import Enum, unique
import validators
import json
```

```python
class Error(Exception):
    """Base class for exceptions in this module."""
    pass

class THREADProcessError(Error):
    """Raised when an operation attempts a state transition that's not
    allowed.

    Attributes:
        message -- explanation of why the specific transition is not allowed
    """

    def __init__(self, message):
        self.message = message
    

class RESTProcessError(Error):
    """Raised when an operation attempts a state transition that's not
    allowed.

    Attributes:
        message -- explanation of why the specific transition is not allowed
    """

    def __init__(self, message):
        self.message = message
```

```python
@unique
class RESTMethod(Enum):
    GET = "GET"
    PUT = "PUT"
    POST = "POST"
    DELETE = "DELETE"
```

```python
@unique
class RESTBodyType(Enum):
    NONE = "none"
    FORM = "form-data"
    XFORM = "x-www-form-data"
    RAW = "raw"
```

```python
@unique
class RESTRawType(Enum):
    TEXT = "text"
    JSON = "json"
```

```python
class RESTConfig(object):
    ID = ""
    URL = ""
    METHOD = ""
    HEADERS = dict()
    BODY_TYPE = ""
    BODY_FORM = dict()
    BODY_XFORM = dict()
    BODY_RAW_TYPE = ""
    BODY_RAW_TEXT = ""
    BODY_RAW_JSON = {}
```

```python
class RESTProcessor(object):
    
    def __init__(self, obj: RestConfig):
        self.CONFIGURATION = obj
        self.logs = []
        
    ## validates the configuraion provided for the REST Processing...
    def validate_configuration(self):
                
        ## URL Verification
        if self.CONFIGURATION.URL == "" or not validators.url(self.CONFIGURATION.URL) :
            raise RESTProcessError("Invalid URL")
            
        ## METHOD verfication
        if self.CONFIGURATION.METHOD == "":
            raise RESTProcessError("Invalid METHOD")
            
        try:
            RESTMethod[self.CONFIGURATION.METHOD]
        except KeyError:
            raise RESTProcessError("Invalid METHOD")
        
    ## headers    
    def add_headers(self):
        
        if self.CONFIGURATION.HEADERS:
            self.HEADERS = self.CONFIGURATION.HEADERS
    
    ## BODY
    def add_body(self):
        
        if self.CONFIGURATION.BODY_TYPE == "":
            raise RESTProcessError("Invalid Body Type")
        
        try:
            RESTBodyType[self.CONFIGURATION.BODY_TYPE]
        except KeyError:
            raise RESTProcessError("Invalid BODY_TYPE")
            
        if self.CONFIGURATION.BODY_TYPE == RESTBodyType.NONE:
            print("None Body Type")
        
        if self.CONFIGURATION.BODY_TYPE == RESTBodyType.FORM:
            self.HEADERS["content-type"] = "application/form-data"
        
        if self.CONFIGURATION.BODY_TYPE == RESTBodyType.XFORM:
            self.HEADERS["content-type"] = "application/x-www-form-data"
        
        if self.CONFIGURATION.BODY_TYPE == RESTBodyType.RAW:
            
            if self.CONFIGURATION.BODY_RAW_TYPE == RESTRawType.TEXT:
                self.HEADERS["content-type"] = "application/text"
            
            if self.CONFIGURATION.BODY_RAW_TYPE == RESTRawType.JSON:
                self.HEADERS["content-type"] = "application/json"
    
    ## add certificates:    
    def add_certificates(self):
        pass
    
    ## process request
    def process_request(self):
        pass
    
```

## Configuration Templates

```python
class RouterConfiguration:
    ID = ""
    REQ = {}
    RES = {}
    THREADS = []

class THREADConfiguration:
    ID = ""
    OUTPUT = ""
    APIS = []    

class MEMORYModel(object):
    
    def __init__(self, id="", typ="", mem=[], pid="" ):
        self.ID = id
        self.TYP = typ
        self.MEMORY = mem
        self.PARENT_ID = pid
    
    def process(self):
        return {
            "id" : self.ID,
            "typ" : self.TYP,
            "objects" : self.MEMORY,
            "parent_id": self.PARENT_ID
        }
```

```python
class MEMORYModeler(object):
    
    def __init__(self, config:RouterConfiguration, id="", typ=""):
        self.id = id
        self.typ = typ
        self.config = config
        self.results = []
        print("Configuration", self.config)
    
    def process(self):
        if self.typ == "ROUTE":
            self.__process_route()
            
        if self.typ == "THREAD":
            self.__process_thread()
        
        if self.typ == "API":
            self.__process_api()
        
    
        print("*" * 100)
        print(self.results)
    
    def __process_route(self):
        
        self.results.append(MEMORYModel(self.id, "route", ["req"], self.id).process())
        
        for t in self.config["threads"]:
            self.results.append(MEMORYModel(t["id"], "thread", ["data"], self.id).process())
            
            for a in t["apis"]:
                self.results.append(MEMORYModel(a["id"], "api", ["req", "res"], t["id"]).process())
        
            
    
    def __process_thread(self):
        
        self.results.append(MEMORYModel(self.config["id"], "route", ["req"], self.config["id"]).process())
        
        for i, t in enumerate(self.config["threads"]):
                                
            if t["id"] == self.id and i == 0:                              
                for a in t["apis"]:
                    self.results.append(MEMORYModel(a["id"], "api", ["req", "res"], t["id"]).process())                
                break
            
            if t["id"] == self.id and i != 0:
                for a in t["apis"]:
                    self.results.append(MEMORYModel(a["id"], "api", ["req", "res"], t["id"]).process())
                break
            
            if t["id"] != self.id and i == 0:
                self.results.append(MEMORYModel(t["id"], "thread", ["data"], self.config["id"]).process())
                
                for a in t["apis"]:
                    self.results.append(MEMORYModel(a["id"], "api", ["req", "res"], t["id"]).process())
                

    
    def __process_api(self):        
        
        self.results.append(MEMORYModel(self.config["id"], "route", ["req"], self.config["id"]).process())
        
        for i, t in enumerate(self.config["threads"]):
            if any(x for x in t["apis"] if x["id"] == self.id):
                break
            
            self.results.append(MEMORYModel(t["id"], "thread", ["data"], self.config["id"]).process())
                
            for a in t["apis"]:
                self.results.append(MEMORYModel(a["id"], "api", ["req", "res"], t["id"]).process())
                
                
                
        
         
```
